```python
# Requires the folder Point3D


Vitesses_dico ={}
for r in [7.5,15,25]:#[7.5,15,25]:
    for v in [5,10,20,40]: #[5,10,20,40]:

        Xstring= "Point3D\\Point3D_R"+ str(r)+"_v"+str(int(v))+".csv"
        dicostring="R"+ str(r)+"_v"+str(int(v))
        
        df=pd.read_csv(Xstring, sep=',', encoding='utf-8',header=None)
        df.rename(columns={0: 'axeZ', 1: 'axeY', 2: 'axeX'}, inplace=True)
        vect_X =df[[c for c in df.columns if c in [5+k*3 for k in range(len(df.columns))]]].astype(float)
        vect_Y =df[[c for c in df.columns if c in [4+k*3 for k in range(len(df.columns))]]].astype(float) ##3 premières colonnes comptent pas, c'est les coordonnées des points, on a bien 303 noramlement
        vect_Z =df[[c for c in df.columns if c in [6+k*3 for k in range(len(df.columns))]]].astype(float)
        #coords=df.loc[:, ['axeZ', 'axeY', 'axeX']] .astype(float)
        
        Vitesses_dico["vect_X_"+dicostring]=vect_X
        Vitesses_dico["vect_Y_"+dicostring]=vect_Y
        Vitesses_dico["vect_Z_"+dicostring]=vect_Z
    
    Vitesses_dico["coords_Z_R"+str(r)]= df.loc[:, ['axeZ']].astype(float)
Vitesses_dico["coords_Y"]= df.loc[:, ['axeY']].astype(float)
Vitesses_dico["coords_X"]= df.loc[:, ['axeX']].astype(float)
```

```python
np.save('Vitesses_Dico.npy', Vitesses_dico) 
```
